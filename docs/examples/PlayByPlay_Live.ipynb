# Working with live PlayByPlay data...

In order to work with live data, you'll need to use a combination of data from nba_api.live and nba_api.stats.
 


```python
# Query nba.live.endpoints for the play by play of GameID 002200011 = IND vs NYK
from nba_api.live.endpoints import playbyplay
actions = playbyplay.PlayByPlay('0022000011').get_dict()
print (actions)

```

```python
# Query nba.live.endpoints for the game Odds of GameID 002200011 = IND vs NYK)
from nba_api.live.endpoints import gameodds
data = gameodds.GameOdds('0022000020').get_dict()
print(data)
```

```python
# Query nba.live.endpoints for the box score of GameID 002200011 = IND vs NYK
import json
from nba_api.live.endpoints import boxscore
data = boxscore.BoxScore('0022000011').get_dict()
print(json.dumps(data))
```

```python
# Query nba.live.endpoints for the score board of GameID 002200011 = IND vs NYK
from nba_api.live.endpoints import scoreboard
data = scoreboard.ScoreBoard().get_dict()
print(data)
```
